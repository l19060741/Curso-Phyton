5. Construye 3 series con 7 elementos cada una y construye un DataFrame. El data frame deberá tener en el axis=0 los 7 elementos y en el axis=1 cada serie.

Nota:
- Prueba hacerlo usando un diccionario.
- Explora el método concat [https://pandas.pydata.org/docs/reference/api/pandas.concat.html]
- Haz dos celdas, una con cada método
- En ambos casos, nombra a las columnas como 'ser1', 'ser2', 'ser3'.
- Explora el atributo df.columns.
- El index será el por default, enteros del 0 al 7

```python
np.random.seed(0)
serie_uno = pd.Series(np.random.rand(7))
serie_dos = pd.Series(np.random.rand(7))
serie_tres = pd.Series(np.random.rand(7))

df = {'ser1':serie_uno, 'ser2':serie_dos, 'ser3':serie_tres}
pd.DataFrame(df)
```

```python

df1 = pd.DataFrame (serie_uno, columns = ['ser1'])
df2 = pd.DataFrame (serie_dos, columns = ['ser2'])
df3 = pd.DataFrame (serie_tres, columns = ['ser3'])
pd.concat([df1, df2, df3], axis = 1)
```

6. Construye un DataFrame con las indicaciones del ejercicio anterior, pero ahora el index serán las letras de la 'a' a  la 'g'

NOTA:
- Si tienes dudas puedes checar este artículo [https://www.geeksforgeeks.org/creating-a-dataframe-from-pandas-series/

```python
np.random.seed(0)
serie_uno = pd.Series(np.random.rand(7),index ='a b c d e f g'.split())
serie_dos = pd.Series(np.random.rand(7),index ='a b c d e f g'.split())
serie_tres = pd.Series(np.random.rand(7),index ='a b c d e f g'.split())

df1 = pd.DataFrame(serie_uno,index ='a b c d e f g'.split(),columns = ['0'])
df2 = pd.DataFrame(serie_dos,index ='a b c d e f g'.split(),columns = ['1'])
df3 = pd.DataFrame(serie_tres,index ='a b c d e f g'.split(),columns = ['2'])

pd.concat([df1, df2, df3], axis = 1)
```
